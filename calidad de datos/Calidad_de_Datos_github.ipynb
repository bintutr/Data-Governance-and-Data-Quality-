```python
import pandas as pd
archivo = pd.read_csv("https://raw.githubusercontent.com/bintutr/Data-Governance-and-Data-Quality-/refs/heads/main/AB_NYC_2019.csv")
for column in archivo.columns:
    null_count = archivo[column].isnull().sum()
    not_null_count = archivo[column].notnull().sum()
    print(f"Columna '{column}': {null_count} valaores nulos, {not_null_count} valores no nulos")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
archivo = pd.read_csv("https://raw.githubusercontent.com/bintutr/Data-Governance-and-Data-Quality-/refs/heads/main/AB_NYC_2019.csv", sep=",")
plt.figure(figsize=(15, 10))

# Scatter plot de 'price' vs 'minimum_nights'
plt.subplot(2, 3, 1)
sns.scatterplot(x=archivo['minimum_nights'], y=archivo['price'])
plt.title('Price vs Minimum Nights')
plt.xlabel('Minimum Nights')
plt.ylabel('Price')

# Scatter plot de 'price' vs 'number_of_reviews'
plt.subplot(2, 3, 2)
sns.scatterplot(x=archivo['number_of_reviews'], y=archivo['price'])
plt.title('Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')

# Scatter plot de 'price' vs 'reviews_per_month'
plt.subplot(2, 3, 3)
sns.scatterplot(x=archivo['reviews_per_month'], y=archivo['price'])
plt.title('Price vs Reviews per Month')
plt.xlabel('Reviews per Month')
plt.ylabel('Price')

# Scatter plot de 'price' vs 'availability_365'
plt.subplot(2, 3, 4)
sns.scatterplot(x=archivo['availability_365'], y=archivo['price'])
plt.title('Price vs Availability (365 days)')
plt.xlabel('Availability (365 days)')
plt.ylabel('Price')

#para visualizar la distribuci√≥n geogr√°fica)
plt.subplot(2, 3, 5)
sns.scatterplot(x=archivo['longitude'], y=archivo['latitude'])
plt.title('Latitude vs Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

plt.tight_layout()
plt.show()
```

**Price vs Minimum Nights:** La mayor√≠a de los alojamientos requieren una estancia m√≠nima corta, pero algunos con estancias m√≠nimas largas tienden a tener precios m√°s altos.

**Price vs Number of Reviews:** Las propiedades con precios moderados tienden a recibir m√°s reviews. Las propiedades de lujo (con precios altos) probablemente sean menos comunes y, por lo tanto, tienen menos reviews.

**Price vs Reviews per Month:** Los alojamientos con precios moderados tienen una tasa m√°s alta de reviews mensuales. Los precios muy altos, de nuevo, muestran menos actividad mensual.

**Price vs Availability (365 days)**: Los alojamientos con precios moderados muestran mayor disponibilidad. Las propiedades con precios extremos (muy altos) tienden a tener menor disponibilidad, posiblemente debido a menos reservas.

**Latitude vs Longitude:** Este gr√°fico muestra la concentraci√≥n de propiedades en diferentes barrios.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

archivo = pd.read_csv("https://raw.githubusercontent.com/bintutr/Data-Governance-and-Data-Quality-/refs/heads/main/AB_NYC_2019.csv", sep=",")
columnas_relevantes = ['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365']
plt.figure(figsize=(15, 10))
for i, col in enumerate(columnas_relevantes, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=archivo[col])
    plt.title(f'Boxplot de {col}')

plt.tight_layout()
plt.show()
```

**Boxplot de price:** La mayor√≠a de los precios est√°n concentrados en el rango m√°s bajo, pero hay algunos valores at√≠picos que llegan hasta los 10,000.

**Boxplot de minimum_nights:** La mayor√≠a de las estancias m√≠nimas est√°n concentradas en el rango bajo, pero hay algunos valores at√≠picos que llegan hasta 1,200 noches.

**Boxplot de number_of_reviews:** La mayor√≠a de las propiedades tienen pocas rese√±as, con algunos outliers que llegan hasta 600 rese√±as.

**Boxplot de reviews_per_month:** La mayor√≠a de las propiedades tienen pocas rese√±as por mes, con algunos outliers que llegan hasta 60 rese√±as por mes.

**Boxplot de availability_365:** La disponibilidad var√≠a ampliamente, con un rango que va desde 0 hasta 365 d√≠as.
